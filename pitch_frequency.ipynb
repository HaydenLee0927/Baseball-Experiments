```python
import numpy as np
import pandas as pd
!pip install pybaseball
from pybaseball import statcast
from pybaseball import playerid_lookup
from pybaseball import statcast_pitcher
import matplotlib.pyplot as plt
import seaborn as sns
```

##Experimenting with Pybaseball

```python
playerid_lookup('judge')
```

```python
kershaw_data = statcast_pitcher('2024-01-01', '2024-12-31', 477132)
kershaw_data.head()
```

```python
plt.hist(kershaw_data["release_speed"], bins = np.arange(70,100,0.5), density = True)
plt.xlabel("release speed")
plt.ylabel("count")
plt.title("release speed histogram")
plt.show()
```

##Experimenting with Seaborn

```python
sns.set_style("whitegrid")
sns.histplot(kershaw_data["release_speed"], bins = np.arange(70,100,0.5), kde = True)
```

```python
sns.kdeplot(kershaw_data["release_speed"])
```

```python

sns.kdeplot(kershaw_data, x = "release_speed", hue = "pitch_type", fill = True)
sns.kdeplot(kershaw_data["release_speed"], linestyle = '--')
```
